```python
cat_data = data[object_cols]

test_cat_data = test[object_cols]
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

cat_data['categoryA'] = label_encoder.fit_transform(cat_data['categoryA'])
cat_data['categoryC'] = label_encoder.fit_transform(cat_data['categoryC'])
cat_data['categoryE'] = label_encoder.fit_transform(cat_data['categoryE'])
cat_data['unit'] = label_encoder.fit_transform(cat_data['unit'])

test_cat_data['categoryA'] = label_encoder.fit_transform(test_cat_data['categoryA'])
test_cat_data['categoryC'] = label_encoder.fit_transform(test_cat_data['categoryC'])
test_cat_data['categoryE'] = label_encoder.fit_transform(test_cat_data['categoryE'])
test_cat_data['unit'] = label_encoder.fit_transform(test_cat_data['unit'])
```

```python
cat_data = pd.get_dummies(cat_data,columns=['categoryB','categoryD','categoryF'])

test_cat_data = pd.get_dummies(test_cat_data,columns=['categoryB','categoryD','categoryF'])
```

```python
new_data = pd.concat([num_data,cat_data],axis=1)

test_new_data = pd.concat([test_num_data,test_cat_data],axis=1)
```

```python
null_columns = new_data.columns[new_data.isnull().any()]

test_null_columns = test_new_data.columns[test_new_data.isnull().any()]
```

```python
for col in null_columns:
    new_data[col].fillna(value=new_data[col].mean(), inplace=True)
    
for test_col in test_null_columns:
    test_new_data[test_col].fillna(value=test_new_data[test_col].mean(), inplace=True)
```

```python
# correleation
corr_new_data = pd.concat([new_data,y],axis=1)
```

```python

features = new_data.columns.tolist()
target = y.name

```

```python
from scipy.stats import pearsonr
correlations = {}
for f in features:
    data_temp = corr_new_data[[f,target]]
    x1 = data_temp[f].values
    x2 = data_temp[target].values
    key = f + ' vs ' + target
    correlations[key] = pearsonr(x1,x2)[0]
```

```python
data_correlations = pd.DataFrame(correlations, index=['Value']).T
data_correlations.loc[data_correlations['Value'].abs().sort_values(ascending=False).index]
```
